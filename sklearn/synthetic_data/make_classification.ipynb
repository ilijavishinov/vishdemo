```python
from sklearn.datasets import make_classification
import seaborn as sns
import importlib
import pandas as pd
```

```python
X, y = make_classification(n_samples = 500,
                           n_features = 2,
                           n_redundant = 0,
                           n_repeated = 0,
                           n_informative = 2,
                           random_state = 42,
                           n_clusters_per_class = 2,
                           n_classes = 2,
                           weights = None)
```

```python
dataset = pd.DataFrame({'feature_1': X[:,0],
                        'feature_2': X[:,1],
                        'target_class': y})
```

```python
sns.scatterplot(data=dataset,
                x = 'feature_1',
                y = 'feature_2',
                hue = 'target_class')
```
